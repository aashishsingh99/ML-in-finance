```python
X=dataset.iloc[:,1:5].values
y=dataset.iloc[:,-1].values
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
```

```python
kwargs=dict(alpha=0.5,bins=100,density=True,stacked=True)
plt.hist(y,**kwargs,color='r',label='true');
plt.hist(y_pred,**kwargs,color='g',label='predicted');
plt.gca().set(title='comparision',ylabel='true and predicted values');
plt.xlim(0,1400)
plt.legend();
```

```python
a = pd.read_csv(r"C:\Users\Aashish Singh\Desktop\ASHOKA.csv", parse_dates=True, index_col='Date',)
b = pd.read_csv(r"C:\Users\Aashish Singh\Desktop\NIFTY50.csv", parse_dates=True, index_col='Date')

monthly1=a.resample('M',how='mean')
monthly2=b.resample('M',how='mean')
monthly_prices = pd.concat([monthly1['Close Price'], monthly2['Close']], axis=1)
monthly_prices.columns = ['ASHOKA', 'NIFTY50']
```

```python
monthly_returns = monthly_prices.pct_change(1)
clean_monthly_returns = monthly_returns.dropna(axis=0)  # drop first missing row
print(clean_monthly_returns.head())
```

```python

import statsmodels.api as sm
X = clean_monthly_returns['NIFTY50']
y = clean_monthly_returns['ASHOKA']
X1 = sm.add_constant(X)
model = sm.OLS(y, X1)
results = model.fit()
print(results.summary())
```

```python
monthly1=a.last('3M')
monthly2=b.last('3M')
monthly_prices = pd.concat([monthly1['Close Price'], monthly2['Close']], axis=1)
monthly_prices.columns = ['ASHOKA', 'NIFTY50']
monthly_returns = monthly_prices.pct_change(1)
clean_monthly_returns = monthly_returns.dropna(axis=0)  # drop first missing row
```

```python

import statsmodels.api as sm
X = clean_monthly_returns['NIFTY50']
y = clean_monthly_returns['ASHOKA']
X1 = sm.add_constant(X)
model = sm.OLS(y, X1)
results = model.fit()
print(results.summary())
```
