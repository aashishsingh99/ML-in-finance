```python
dfnew=pd.read_csv(r"C:\Users\Aashish Singh\Desktop\ASHOKA.csv")
dfnew=dfnew[dfnew['Series']=='EQ']
a=dfnew['Average Price']
dfnew['21']=a.rolling(window=21).mean()
dfnew['34']=a.rolling(window=34).mean()
dfnew['21'].plot()
dfnew['34'].plot()
for index, row in dfnew.iterrows():
    if(row['21']>row['34']):
        dfnew['Trade']='buy'
    elif(row['21']<row['34']):
        dfnew['Trade']='sell'
dfnew
```

```python
a=dfnew['Close Price']
dfnew['sma'] = a.rolling(window=14).mean()
dfnew['rstd'] = a.rolling(window=14).std()
dfnew['upper_band'] = sma + 2 * rstd
dfnew['lower_band'] = sma - 2 * rstd
dfnew.plot(x='Date',y=['Close Price','sma','upper_band','lower_band'])
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
x_axis = dfnew.index.get_level_values(0)
ax.fill_between(x_axis, dfnew['upper_band'], dfnew['lower_band'], color='grey')
ax.plot(x_axis, dfnew['Average Price'], color='blue', lw=2)
ax.plot(x_axis, dfnew['sma'], color='black', lw=2)
dfnew.to_csv(r"C:\Users\Aashish Singh\Desktop\new.csv")
```
