```python
df_filtered.rename(columns={'Close Price':'close'},inplace=True)
df_filtered['Day_Perc_Change'] = (df_filtered.close.pct_change(periods = 1))*100
df_filtered.rename(columns={'close':'Close Price'},inplace=True)
```

```python
def func3(row):
    if (row['Day_Perc_Change']<-7):
        return 'Bear Drop'
    elif (row['Day_Perc_Change']> -7 and row['Day_Perc_Change'] <-3):
        return 'Among Top Losers'
    elif (row['Day_Perc_Change']> -3 and row['Day_Perc_Change'] <-1):
        return 'Negative'
    elif (row['Day_Perc_Change']> -0.5 and row['Day_Perc_Change'] <0.5):
        return 'Slight or No change'
    elif (row['Day_Perc_Change']> 0.5 and row['Day_Perc_Change'] <1):
        return 'Slight positive'
    elif (row['Day_Perc_Change']> 1 and row['Day_Perc_Change'] <3):
        return 'Positive'
    
    elif (row['Day_Perc_Change']> 3 and row['Day_Perc_Change'] <7):
        return 'Among Top Gainers'
    
    elif (row['Day_Perc_Change']> 7):
        return 'Bull Run'
    
df_filtered['Trend'] = df_filtered.apply(lambda row: func3(row), axis=1)
```

```python


df_filtered.groupby(df_filtered.Trend).mean()['Total Traded Quantity']
df_filtered.groupby(df_filtered.Trend).median()['Total Traded Quantity']
df_filtered
```
