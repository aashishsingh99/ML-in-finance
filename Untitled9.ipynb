```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Aashish Singh\Desktop\new.csv")
df
```

```python
df.loc[df['Close Price']<df['lower_band'],'Call']='buy'
df.loc[(df['Close Price']>df['lower_band']) & (df['Close Price']<df['sma']),'Call']='Hold Buy/ Liquidate Short'
df.loc[(df['Close Price']>df['sma']) & (df['Close Price']<df['upper_band']),'Call']='Hold Buy/ Liquidate Buy'
df.loc[df['Close Price']>df['upper_band'],'Call']='Short'
df = df[np.isfinite(df['sma'])]
```

```python
#Logistic regression
X=df.iloc[:,[9,19,21,22]].values
y=df.iloc[:,-1].values
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm=np.asarray(cm)
accuracy=np.trace(cm)/np.sum(cm)
print(accuracy*100)
```

```python
#K-nearest neighbours
X=df.iloc[:,[9,19,21,22]].values
y=df.iloc[:,-1].values
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm=np.asarray(cm)
accuracy=np.trace(cm)/np.sum(cm)
print(accuracy*100)
```

```python
#SVM
X=df.iloc[:,[9,19,21,22]].values
y=df.iloc[:,-1].values
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
from sklearn.svm import SVC
classifier1=SVC(kernel='linear')
classifier1.fit(X_train,y_train)
y_pred=classifier1.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm=np.asarray(cm)
accuracy=np.trace(cm)/np.sum(cm)
print(accuracy*100)
```

```python
#Decision tree classifier
X=df.iloc[:,[9,19,21,22]].values
y=df.iloc[:,-1].values
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy')
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm=np.asarray(cm)
accuracy=np.trace(cm)/np.sum(cm)
print(accuracy*100)
```

```python
#Random Forest classifier

X=df.iloc[:,[9,19,21,22]].values
y=df.iloc[:,-1].values
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy')
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm=np.asarray(cm)
accuracy=np.trace(cm)/np.sum(cm)
print(accuracy*100)
```

```python
#naive bayes classifier

X=df.iloc[:,[9,19,21,22]].values
y=df.iloc[:,-1].values
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm=np.asarray(cm)
accuracy=np.trace(cm)/np.sum(cm)
print(accuracy*100)
```

```python
df=pd.read_csv(r"C:\Users\Aashish Singh\Desktop\BAJAJELEC.csv")
df['o_c_perc']=((df['Open Price']-df['Close Price'])/(df['Open Price']))*100
df['l_h_perc']=((df['Low Price']-df['High Price'])/(df['Low Price']))*100
df['rolling_mean'] = df['Close Price'].pct_change().rolling(5).mean()
df['rolling_std'] = df['Close Price'].pct_change().rolling(5).std()
df
```

```python
df['P'] = df['Close Price'].shift(-1)
df
```

```python
df['Action']=np.nan
for index, row in df.iterrows():
    if (row.P > row['Close Price']):
        df.at[index, 'Action'] = 1
    elif(row.P < row['Close Price']):
        df.at[index, 'Action'] = -1
df=df.drop('P',1)
df
```

```python

df = df[np.isfinite(df['rolling_mean'])]
df = df[np.isfinite(df['Action'])]
X=df.iloc[:,[-5,-4,-3,-2]].values
y=df.iloc[:,-1].values
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy')
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm=np.asarray(cm)
accuracy=np.trace(cm)/np.sum(cm)
print(accuracy*100)
df
```
